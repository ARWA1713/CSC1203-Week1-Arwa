```python
import math

# Function to calculate area
def calculate_area(radius):
    return math.pi * radius ** 2

# Lambda function to calculate circumference
calculate_circumference = lambda radius: 2 * math.pi * radius

# User input
radius = float(input("Enter the radius of the circle: "))

# Calculations
area = calculate_area(radius)
circumference = calculate_circumference(radius)

# Output rounded to 2 decimal places
print("Area:", round(area, 2))
print("Circumference:", round(circumference, 2))
```

```python
import re

# Function to sum numbers
def sum_numbers(numbers):
    return sum(numbers)

# Lambda to double the sum
double_value = lambda x: x * 2

# User input
text = input("Enter a string: ")

# Find all numbers
numbers_found = re.findall(r'\d+', text)

# Convert to integers
numbers = list(map(int, numbers_found))

# Calculations
total = sum_numbers(numbers)
doubled = double_value(total)

# Output
print("Numbers found:", numbers)
print("Sum:", total)
print("Doubled value:", doubled)
```

```python
import math

# Function to calculate squares using lambda
def calculate_squares(numbers):
    return list(map(lambda x: x ** 2, numbers))

# User input
user_input = input("Enter numbers separated by commas: ")

# Convert to list of floats
numbers = list(map(float, user_input.split(",")))

# Calculate squares
squares = calculate_squares(numbers)

# Math calculations
max_square = max(squares)
sqrt_sum_squares = math.sqrt(sum(squares))

# Output
print("Squares:", squares)
print("Maximum square:", max_square)
print("Square root of sum of squares:", round(sqrt_sum_squares, 2))
```

```python
import re
import math

# Function to calculate average using lambda
def calculate_average(scores):
    avg = (lambda s: sum(s) / len(s))(scores)
    return avg

# Password input
password = input("Enter password: ")

# Regex checks
has_upper = re.search(r'[A-Z]', password)
has_lower = re.search(r'[a-z]', password)
has_digit = re.search(r'\d', password)

# Check strength
if has_upper and has_lower and has_digit:
    strength = "Strong"
else:
    strength = "Weak"

# Scores input
scores_input = input("Enter 5 scores separated by spaces: ")
scores = list(map(float, scores_input.split()))

average = calculate_average(scores)
rounded_average = math.ceil(average)

# Output
print("Password is:", strength)
print("Rounded Average Score:", rounded_average)
```

```python
import re
import math

# Function to count total vowels
def count_vowels(text):
    vowels = re.findall(r'[aeiouAEIOU]', text)
    return len(vowels)

# User input
paragraph = input("Enter a paragraph: ")

# Find words starting with a vowel
vowel_words = re.findall(r'\b[aeiouAEIOU]\w*', paragraph)

# Lambda to calculate word lengths
word_lengths = list(map(lambda word: len(word), vowel_words))

# Calculations
total_vowels = count_vowels(paragraph)

if word_lengths:
    average_length = round(sum(word_lengths) / len(word_lengths), 2)
else:
    average_length = 0

# Output
print("Words starting with vowels:", vowel_words)
print("Total vowels in paragraph:", total_vowels)
print("Average vowel-word length:", average_length)
```
