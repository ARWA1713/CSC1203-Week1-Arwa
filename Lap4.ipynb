```python
total = 0

for i in range(10):
    num = int(input("Enter a number: "))

    if num == 0:
        break

    if num < 0:
        continue

    total += num

print("Final Sum =", total)
```

```python
num = int(input("Enter a number: "))

for i in range(1, 13):
    product = num * i

    if product > 50:
        break

    if product % 3 == 0:
        continue

    print(f"{num} x {i} = {product}")
```

```python
secret_number = 17  
attempts = 0

while True:
    guess = int(input("Guess a number between 1 and 20: "))

    if guess < 1 or guess > 20:
        print("Invalid guess! Try again.")
        continue

    attempts += 1

    if guess == secret_number:
        print("Correct! ")
        break
    else:
        print("Wrong guess!")

print("Number of valid attempts:", attempts)
```

```python
n = int(input("Enter a number: "))
count = 0

for i in range(1, n + 1):

    if i > 50:
        break

    if i % 8 == 0:
        continue

    if i % 4 == 0:
        count += 1

print("Count =", count)
```
