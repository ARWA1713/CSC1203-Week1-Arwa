```python
try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    result = num1 / num2
except ZeroDivisionError:
    print("Error: Division by zero is not allowed")
except ValueError:
    print("Error: Invalid input")
else:
    print("Result =", result)
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

class Manager(Employee):
    def add_bonus(self, bonus):
        self.salary = self.salary + bonus

    def display_info(self):
        print("Name:", self.name)
        print("Total Salary:", self.salary)

manager1 = Manager("Arwa", 5000)

manager1.add_bonus(1000)

manager1.display_info()
```

```python
# Program 2: Person Class
class Person:
    def __init__(self, name, age):
        self.name = name          # public property
        self._age = age           # private property (using _)
    
    def update_age(self, new_age):
        self._age = new_age
    
    def display_details(self):
        print(f"Name: {self.name}, Age: {self._age}")

# Create object
person1 = Person("Arwa", 25)
print("After update:")
person1.update_age(26)
person1.display_details()
```

```python
# Program 4: Student Class with Exception Handling
class Student:
    def __init__(self):
        try:
            self.name = input("Enter student name: ")
            self.age = int(input("Enter student age: "))
            if self.age < 0:
                print("Error: Age cannot be negative")
                self.age = 0
        except ValueError:
            print("Error: Please enter a valid number for age")
            self.age = 0
    
    def display_details(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Create object and display details
student1 = Student()
student1.display_details()
```
